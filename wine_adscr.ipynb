Advanced Descriptive Statistics Tutorial Using the Wine Dataset in Python

In this tutorial, we will dive deeper into descriptive statistics using the Wine dataset. 

We will explore grouping by the target variable (wine class) and analyze specific features in more detail. This will help us understand how the chemical properties of wine vary across different classes and identify patterns or trends.

We will use Python libraries such as pandas, numpy, matplotlib, and seaborn for this analysis.

Import Libraries

First, let's import the necessary libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

Load the Wine Dataset
    
The Wine dataset is available in the scikit-learn library. Let's load it and convert it into a pandas DataFrame.

```python
# Load the Wine dataset
wine = load_wine()

# Convert to pandas DataFrame
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Add the target column to the DataFrame
df['target'] = wine.target

# Display the first few rows of the dataset
print(df.head())
```

Grouping by Target Variable

The target variable (target) represents the wine class (0, 1, or 2). 

Let's group the data by the target variable and analyze the descriptive statistics for each class.

Grouped Summary Statistics

We can calculate summary statistics (e.g., mean, median, standard deviation) for each feature, grouped by the target variable.

```python
# Group by target variable and calculate mean
grouped_mean = df.groupby('target').mean()
print("Mean values by wine class:\n", grouped_mean)
```

```python
# Group by target variable and calculate median
grouped_median = df.groupby('target').median()
print("Median values by wine class:\n", grouped_median)
```

```python
# Group by target variable and calculate standard deviation
grouped_std = df.groupby('target').std()
print("Standard deviation by wine class:\n", grouped_std)
```

Visualizing Grouped Statistics

Let's visualize the mean values of some key features across the wine classes.

```python
# Plot mean alcohol content by wine class
plt.figure(figsize=(8, 6))
sns.barplot(x=grouped_mean.index, y=grouped_mean['alcohol'], palette='viridis')
plt.title('Mean Alcohol Content by Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Mean Alcohol Content')
plt.show()
```

```python
# Plot mean malic acid by wine class
plt.figure(figsize=(8, 6))
sns.barplot(x=grouped_mean.index, y=grouped_mean['malic_acid'], palette='viridis')
plt.title('Mean Malic Acid by Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Mean Malic Acid')
plt.show()
```

Analyzing Specific Features

Let's analyze specific features in more detail to understand their distribution and relationship with the target variable.

Alcohol Content

Alcohol content is a key feature in wine analysis. Let's explore its distribution and relationship with the target variable.

```python
# Plot alcohol content distribution by wine class
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='alcohol', data=df, palette='viridis')
plt.title('Alcohol Content Distribution by Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
# Plot alcohol content histogram for each wine class
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='alcohol', hue='target', kde=True, palette='viridis')
plt.title('Alcohol Content Histogram by Wine Class')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

Malic Acid

Malic acid is another important feature. Let's analyze its distribution and relationship with the target variable.

```python
# Plot malic acid distribution by wine class
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='malic_acid', data=df, palette='viridis')
plt.title('Malic Acid Distribution by Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Malic Acid')
plt.show()
```

```python
# Plot malic acid histogram for each wine class
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='malic_acid', hue='target', kde=True, palette='viridis')
plt.title('Malic Acid Histogram by Wine Class')
plt.xlabel('Malic Acid')
plt.ylabel('Frequency')
plt.show()
```

Flavonoids

Flavonoids are a group of chemical compounds found in wine. Let's analyze their distribution and relationship with the target variable.


```python
# Plot flavonoids distribution by wine class
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='flavanoids', data=df, palette='viridis')
plt.title('Flavonoids Distribution by Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Flavonoids')
plt.show()
```

```python
# Plot flavonoids histogram for each wine class
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='flavanoids', hue='target', kde=True, palette='viridis')
plt.title('Flavonoids Histogram by Wine Class')
plt.xlabel('Flavonoids')
plt.ylabel('Frequency')
plt.show()
```

Correlation Analysis

Let's analyze the correlation between features and the target variable.

Correlation Matrix

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Dataset Features')
plt.show()
```

Correlation with Target Variable

Let's analyze the correlation of each feature with the target variable.

```python
# Calculate correlation with the target variable
corr_with_target = df.corr()['target'].sort_values(ascending=False)
print("Correlation with Target Variable:\n", corr_with_target)
```

```python
# Plot correlation with the target variable
plt.figure(figsize=(10, 6))
sns.barplot(x=corr_with_target.index, y=corr_with_target.values, palette='viridis')
plt.title('Correlation of Features with Target Variable')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=90)
plt.show()
```
